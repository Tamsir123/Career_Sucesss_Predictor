```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Colonnes à encoder
multi_label_cols = ['Languages_Spoken', 'Certifications']
one_hot_cols = ['Field_of_Study', 'Location', 'Gender', 'Current_Job_Level']

# Sauvegarde des dimensions initiales et affichage des colonnes
print(f"Dimensions avant encodage : {df.shape}")
print("Colonnes avant encodage : ", df.columns.tolist())
print("Échantillon avant encodage :")
print(df[multi_label_cols + one_hot_cols].head())

# Encodage multi-label
for col in multi_label_cols:
    print(f"\nEncodage de {col}...")
    # Remplir les valeurs manquantes et préparer les données
    df[col] = df[col].fillna('')
    split_data = df[col].astype(str).apply(lambda x: [i.strip() for i in x.split(',') if i.strip()])
    print(f"Exemple de données splittées pour {col} : ", split_data.head().tolist())
    
    # Appliquer MultiLabelBinarizer
    mlb = MultiLabelBinarizer()
    encoded = pd.DataFrame(mlb.fit_transform(split_data),
                           columns=[f"{col}_{cls}" for cls in mlb.classes_],
                           index=df.index)
    # Convertir explicitement en int pour éviter True/False
    encoded = encoded.astype('uint8')  # uint8 pour économiser de la mémoire
    print(f"Nouvelles colonnes pour {col} : ", encoded.columns.tolist())
    
    # Concaténer et supprimer la colonne originale
    df = pd.concat([df.drop(columns=[col]), encoded], axis=1)
    print(f"Dimensions après encodage de {col} : {df.shape}")

# Encodage one-hot
print("\nEncodage one-hot...")
# Gérer les valeurs manquantes
for col in one_hot_cols:
    df[col] = df[col].fillna('Unknown')
# Appliquer get_dummies
df = pd.get_dummies(df, columns=one_hot_cols, drop_first=True)
print(f"Dimensions après encodage one-hot : {df.shape}")

# Convertir uniquement les colonnes encodées en uint8 (0 et 1)
encoded_cols = [col for col in df.columns if any(col.startswith(prefix) for prefix in ['Languages_Spoken_', 'Certifications_', 'Field_of_Study_', 'Location_', 'Gender_', 'Current_Job_Level_'])]
df[encoded_cols] = df[encoded_cols].astype('uint8')  # Convertir seulement les colonnes encodées

# Résultat final
print(f"\n✅ Nouvelles dimensions du dataset : {df.shape}")
print("📋 Colonnes encodées :")
print(df.columns.tolist())
print("\nÉchantillon des données encodées :")
print(df.head())
print("\nTypes de données des colonnes :")
print(df.dtypes)
print("\nVérification des valeurs dans une colonne encodée :")
print(df[encoded_cols[:2]].head())  # Affiche les 2 premières colonnes encodées
```
