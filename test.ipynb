```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données
df = pd.read_csv('education_career_success_g.csv')

# Sélectionner uniquement les colonnes numériques
numeric_data = df.select_dtypes(include=['float64', 'int64'])

# Calcul de la matrice de corrélation
correlation_matrix = numeric_data.corr()

# Visualisation de la matrice de corrélation avec une heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap des corrélations')
plt.show()

# Distribution des colonnes numériques
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_data.columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(numeric_data[column], bins=20, kde=True)
    plt.title(f'Distribution de {column}')
plt.tight_layout()
plt.show()

# Distribution des colonnes catégorielles
categorical_data = df.select_dtypes(include=['object'])
n_categorical_cols = len(categorical_data.columns)
n_rows = (n_categorical_cols + 2) // 3  # Arrondir au nombre entier supérieur si nécessaire

plt.figure(figsize=(15, 5 * n_rows))
for i, column in enumerate(categorical_data.columns, 1):
    plt.subplot(n_rows, 3, i)
    sns.countplot(y=column, data=df, order=df[column].value_counts().index)
    plt.title(f'Distribution de {column}')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Colonnes à encoder
multi_label_cols = ['Languages_Spoken', 'Certifications']
one_hot_cols = ['Field_of_Study', 'Location', 'Gender', 'Current_Job_Level']

# Encodage multi-label
for col in multi_label_cols:
    df[col] = df[col].fillna('')
    split_data = df[col].astype(str).apply(lambda x: [i.strip() for i in x.split(',') if i.strip()])
    mlb = MultiLabelBinarizer()
    encoded = pd.DataFrame(mlb.fit_transform(split_data),
                           columns=[f"{col}_{cls}" for cls in mlb.classes_],
                           index=df.index)
    df = pd.concat([df.drop(columns=[col]), encoded], axis=1)

# Encodage one-hot (avec drop_first pour éviter la redondance)
df = pd.get_dummies(df, columns=one_hot_cols, drop_first=True)

# Résultat final
print(f"✅ Nouvelles dimensions du dataset : {df.shape}")
print("📋 Colonnes encodées :")
print(df.columns.tolist())

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MultiLabelBinarizer

correlation_matrix = df.corr()

# Visualisation de la matrice de corrélation avec une heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap des corrélations')
plt.show()

# Sélection des colonnes numériques et des colonnes encodées
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
encoded_cols = [col for col in df.columns if col not in numeric_cols]

# Calcul de la matrice de corrélation pour les colonnes sélectionnées
correlation_matrix_encoded = df[numeric_cols + encoded_cols].corr()

# Visualisation de la matrice de corrélation avec une heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix_encoded, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap des corrélations avec les variables catégorielles encodées')
plt.show()

# Sélection des variables importantes pour l'analyse
important_vars = ['University_GPA', 'Internships_Completed', 'Soft_Skills_Score', 'Networking_Score', 'Field_of_Study', 'Location', 'Gender', 'Starting_Salary']

# Calcul de la matrice de corrélation pour les variables importantes
correlation_matrix_important = df[important_vars].corr()

# Visualisation de la matrice de corrélation avec une heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix_important, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap des corrélations pour les variables importantes')
plt.show()
```
